### What is color Thresholding

Color Thresholding is techinque to filter out a specific range of colors in an image by defining 
an upper limit and lower limti in a color space. we use it segment , replace or isolate specific colors
in an image

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread("../basics_resources_and_outputs/mountain.jpg")
```

```python
image_rgb = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(image_rgb)
```

```python
plt.imshow(image) # cv2 by default color is bgr
```

```python
img_hsv = cv2.cvtColor(image_rgb,cv2.COLOR_BGR2HSV)
```

```python
plt.imshow(img_hsv)
```

```python
lower = np.array([0,0,0])
upper = np.array([240,0,255])
```

```python
mask = cv2.inRange(img_hsv,lower,upper)
```

```python
plt.imshow(mask)
```

```python
black_hsv = np.uint8([[[120,255,0]]])
black_bgr = cv2.cvtColor(black_hsv, cv2.COLOR_HSV2BGR)[0][0]
```

```python
image_result = image.copy()
image_result[mask > 0] = black_bgr
```

```python
image_result_rgb = cv2.cvtColor(image_result,cv2.COLOR_BGR2RGB)
```

```python
fig, axs = plt.subplots(1,3,figsize = (15,7))
axs[0].imshow(image_rgb)
axs[0].axis('off')
axs[0].set_title("Original Image")

axs[1].imshow(mask)
axs[1].axis('off')
axs[1].set_title("Binary Mask")

axs[2].imshow(image_result_rgb)
axs[2].axis('off')
axs[2].set_title("Output Image")

plt.tight_layout()
plt.show()
```
